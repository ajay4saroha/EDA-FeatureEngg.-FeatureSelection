```python
dataset = pd.read_csv('house-prediction.csv')
```

```python
categorical_feature_nan = [ feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype=='O']
```

```python
for feature in categorical_feature_nan:
    print("{} : {}%".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
for feature in categorical_feature_nan:
    dataset[feature]=dataset[feature].fillna("Missing")
```

```python
dataset.head()
```

```python
numerical_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']   
```

```python
numerical_nan
```

```python
for feature in numerical_nan:
    print("{} : {}%".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
for feature in numerical_nan:
    median_value=dataset[feature].median()
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature]=dataset[feature].fillna(median_value,inplace=True)
```

```python
dataset
```

```python
num_feature = [feature for feature in dataset.columns if dataset[feature].isnull().sum()==0 and dataset[feature].dtype!='Object']
num_feature
```

## Handlin rare categorical Feature

```python
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtype=='O']
```

```python
categorical_feature
```

```python
for feature in categorical_feature:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare')
```
